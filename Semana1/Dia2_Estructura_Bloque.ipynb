## 📖 Contenido Teórico

### Anatomía Completa de un Bloque

Un bloque en blockchain contiene varios campos esenciales. Cada uno cumple una función específica:

#### Campos Típicos:

1. **`index`** (Índice)  
   - Número secuencial que indica la posición del bloque en la cadena
   - El bloque génesis tiene index = 0
   - Facilita la navegación y validación de la cadena

2. **`timestamp`** (Marca de tiempo)  
   - Registro exacto del momento de creación del bloque
   - Formato estándar: ISO 8601 o Unix timestamp
   - Permite ordenar eventos cronológicamente

3. **`data`** (Datos/Transacciones)  
   - Contenido real del bloque
   - Puede ser: transacciones financieras, contratos inteligentes, documentos, etc.
   - En blockchains reales, suele ser una lista de transacciones

4. **`previous_hash`** (Hash del bloque anterior)  
   - Referencia criptográfica al bloque previo
   - **Este campo es el que "encadena" los bloques**
   - Si cambia el bloque anterior, este campo se invalida

5. **`hash`** (Hash del bloque actual)  
   - Huella digital única calculada a partir de todos los campos del bloque
   - Se genera usando funciones criptográficas (SHA-256)
   - Cualquier cambio en el bloque produce un hash completamente diferente

6. **`nonce`** (Number used Once)  
   - Número arbitrario que se ajusta durante el proceso de minería
   - Se modifica repetidamente hasta encontrar un hash que cumpla ciertos requisitos
   - Esencial en algoritmos de Proof of Work (PoW)

## 💻 Código de Ejemplo: Bloque con Timestamp Automático

Vamos a crear bloques usando la librería `datetime` de Python para generar timestamps reales.

````python
# Importar la librería datetime para manejar fechas y horas
from datetime import datetime

# Función auxiliar para crear un bloque
def crear_bloque(index, data, previous_hash, nonce=0):
    """
    Crea un bloque con estructura completa.
    
    Parámetros:
    - index: posición del bloque en la cadena
    - data: información a almacenar
    - previous_hash: hash del bloque anterior
    - nonce: número usado en minería (por defecto 0)
    
    Retorna: diccionario con la estructura del bloque
    """
    bloque = {
        'index': index,
        'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),  # Genera timestamp actual
        'data': data,
        'previous_hash': previous_hash,
        'nonce': nonce,
        'hash': 'temporal_hash'  # Por ahora es temporal; lo calcularemos en Día 3
    }
    return bloque

# Crear el bloque génesis
bloque_genesis = crear_bloque(
    index=0,
    data='Bloque Génesis - Inicio de la cadena',
    previous_hash='0',  # No hay bloque anterior
    nonce=0
)

# Mostrar el bloque
print("=" * 60)
print("BLOQUE GÉNESIS")
print("=" * 60)
for campo, valor in bloque_genesis.items():
    print(f"{campo.upper():15} : {valor}")
print("=" * 60)
````

### 🔍 Explicación del Código

1. **`from datetime import datetime`**  
   Importamos la clase `datetime` para trabajar con fechas y horas.

2. **`datetime.now()`**  
   Obtiene la fecha y hora actual del sistema.

3. **`.strftime('%Y-%m-%d %H:%M:%S')`**  
   Formatea el datetime en una cadena legible:  
   - `%Y` = año (4 dígitos)  
   - `%m` = mes (01-12)  
   - `%d` = día (01-31)  
   - `%H:%M:%S` = hora:minuto:segundo

4. **Función `crear_bloque()`**  
   Encapsula la lógica de creación para reutilizar código fácilmente.

5. **Campo `nonce`**  
   Por ahora es 0, pero en minería se ajustaría hasta encontrar un hash válido.

````python
# SOLUCIÓN COMENTADA

from datetime import datetime

# La función ya acepta el parámetro nonce (ver código de ejemplo arriba)
def crear_bloque(index, data, previous_hash, nonce=0):
    bloque = {
        'index': index,
        'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
        'data': data,
        'previous_hash': previous_hash,
        'nonce': nonce,  # Campo nonce agregado
        'hash': 'temporal_hash'
    }
    return bloque

# Crear bloque_1 con nonce específico
bloque_1 = crear_bloque(
    index=1,
    data='Transacción: Alice envía 50 tokens a Bob',
    previous_hash='temporal_hash',  # Hash del bloque génesis
    nonce=12345  # Nonce usado en minería
)

# Mostrar el bloque
print("=" * 60)
print("BLOQUE #1")
print("=" * 60)
for campo, valor in bloque_1.items():
    print(f"{campo.upper():15} : {valor}")
print("=" * 60)

# Respuesta sobre el nonce:
# El nonce (Number used Once) es un número que los mineros modifican
# repetidamente durante el proceso de minería. Su propósito es encontrar
# un hash que cumpla con ciertos requisitos de dificultad (por ejemplo,
# que el hash comience con una cantidad específica de ceros).
# 
# Proceso:
# 1. Minero calcula hash del bloque con nonce=0
# 2. Si el hash no cumple requisitos, incrementa nonce=1 y recalcula
# 3. Repite hasta encontrar un hash válido (puede tomar millones de intentos)
# 4. El nonce ganador se incluye en el bloque como "prueba de trabajo"
#
# Esto hace que crear bloques requiera esfuerzo computacional,
# protegiendo la red contra ataques de spam y alteraciones.
````
