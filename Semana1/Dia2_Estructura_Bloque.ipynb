### AnatomÃ­a Completa de un Bloque

Un bloque en blockchain contiene varios campos esenciales. Cada uno cumple una funciÃ³n especÃ­fica:

#### Campos TÃ­picos:

1. **`index`** (Ãndice)  
   - NÃºmero secuencial que indica la posiciÃ³n del bloque en la cadena
   - El bloque gÃ©nesis tiene index = 0
   - Facilita la navegaciÃ³n y validaciÃ³n de la cadena

2. **`timestamp`** (Marca de tiempo)  
   - Registro exacto del momento de creaciÃ³n del bloque
   - Formato estÃ¡ndar: ISO 8601 o Unix timestamp
   - Permite ordenar eventos cronolÃ³gicamente

3. **`data`** (Datos/Transacciones)  
   - Contenido real del bloque
   - Puede ser: transacciones financieras, contratos inteligentes, documentos, etc.
   - En blockchains reales, suele ser una lista de transacciones

4. **`previous_hash`** (Hash del bloque anterior)  
   - Referencia criptogrÃ¡fica al bloque previo
   - **Este campo es el que "encadena" los bloques**
   - Si cambia el bloque anterior, este campo se invalida

5. **`hash`** (Hash del bloque actual)  
   - Huella digital Ãºnica calculada a partir de todos los campos del bloque
   - Se genera usando funciones criptogrÃ¡ficas (SHA-256)
   - Cualquier cambio en el bloque produce un hash completamente diferente

6. **`nonce`** (Number used Once)  
   - NÃºmero arbitrario que se ajusta durante el proceso de minerÃ­a
   - Se modifica repetidamente hasta encontrar un hash que cumpla ciertos requisitos
   - Esencial en algoritmos de Proof of Work (PoW)

## ğŸ’» CÃ³digo de Ejemplo: Bloque con Timestamp AutomÃ¡tico

Vamos a crear bloques usando la librerÃ­a `datetime` de Python para generar timestamps reales.

````python
# Importar la librerÃ­a datetime para manejar fechas y horas
from datetime import datetime

# FunciÃ³n auxiliar para crear un bloque
def crear_bloque(index, data, previous_hash, nonce=0):
    """
    Crea un bloque con estructura completa.
    
    ParÃ¡metros:
    - index: posiciÃ³n del bloque en la cadena
    - data: informaciÃ³n a almacenar
    - previous_hash: hash del bloque anterior
    - nonce: nÃºmero usado en minerÃ­a (por defecto 0)
    
    Retorna: diccionario con la estructura del bloque
    """
    bloque = {
        'index': index,
        'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),  # Genera timestamp actual
        'data': data,
        'previous_hash': previous_hash,
        'nonce': nonce,
        'hash': 'temporal_hash'  # Por ahora es temporal; lo calcularemos en DÃ­a 3
    }
    return bloque

# Crear el bloque gÃ©nesis
bloque_genesis = crear_bloque(
    index=0,
    data='Bloque GÃ©nesis - Inicio de la cadena',
    previous_hash='0',  # No hay bloque anterior
    nonce=0
)

# Mostrar el bloque
print("=" * 60)
print("BLOQUE GÃ‰NESIS")
print("=" * 60)
for campo, valor in bloque_genesis.items():
    print(f"{campo.upper():15} : {valor}")
print("=" * 60)
````

   Obtiene la fecha y hora actual del sistema.

3. **`.strftime('%Y-%m-%d %H:%M:%S')`**  
   Formatea el datetime en una cadena legible:  
   - `%Y` = aÃ±o (4 dÃ­gitos)  
   - `%m` = mes (01-12)

   - `%H:%M:%S` = hora:minuto:segundo
