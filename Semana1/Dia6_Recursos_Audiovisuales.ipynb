## üî¨ Experimentos Pr√°cticos Adicionales

### Experimento 1: Comparar Funciones Hash

**Objetivo:** Entender diferencias entre algoritmos hash  
**Actividad:**

```python
import hashlib

texto = "Blockchain"

# Diferentes algoritmos hash
hash_md5 = hashlib.md5(texto.encode()).hexdigest()
hash_sha1 = hashlib.sha1(texto.encode()).hexdigest()
hash_sha256 = hashlib.sha256(texto.encode()).hexdigest()
hash_sha512 = hashlib.sha512(texto.encode()).hexdigest()

print("Comparaci√≥n de Algoritmos Hash")
print("=" * 70)
print(f"Texto: {texto}")
print()
print(f"MD5    ({len(hash_md5)} chars):  {hash_md5}")
print(f"SHA1   ({len(hash_sha1)} chars): {hash_sha1}")
print(f"SHA256 ({len(hash_sha256)} chars): {hash_sha256}")
print(f"SHA512 ({len(hash_sha512)} chars): {hash_sha512}")
print("=" * 70)
print("\nüí° Observa: Mayor longitud = Mayor seguridad (pero m√°s lento)")
print("   MD5 y SHA1 est√°n obsoletos para seguridad cr√≠tica.")
print("   SHA256 es el est√°ndar actual en blockchain.")
```

### Experimento 2: Medir Velocidad de Hashing

```python
import hashlib
import time

# Generar un texto grande
texto_grande = "Blockchain " * 100000  # 1.1 MB aproximadamente

# Medir tiempo de hashing
inicio = time.time()
hash_resultado = hashlib.sha256(texto_grande.encode()).hexdigest()
fin = time.time()

tiempo_transcurrido = (fin - inicio) * 1000  # Convertir a milisegundos

print("Prueba de Velocidad de Hashing")
print("=" * 70)
print(f"Tama√±o del texto: {len(texto_grande):,} caracteres")
print(f"Tiempo de hashing: {tiempo_transcurrido:.2f} ms")
print(f"Hash resultante: {hash_resultado[:32]}...")
print("=" * 70)
print("\nüí° Conclusi√≥n: SHA-256 es extremadamente r√°pido,")
print("   incluso para grandes cantidades de datos.")
```
