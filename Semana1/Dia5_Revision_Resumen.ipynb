### ü§î Preguntas de Reflexi√≥n

Antes de continuar, reflexiona sobre estas preguntas:

1. **¬øPor qu√© la blockchain es inmutable?**  
   *Respuesta esperada:* Porque cada bloque contiene el hash del anterior. Alterar un bloque cambia su hash, rompiendo el enlace con el siguiente bloque. Para ocultar la alteraci√≥n, habr√≠a que recalcular todos los bloques posteriores m√°s r√°pido que la red agrega nuevos bloques.

2. **¬øQu√© limita la manipulaci√≥n de blockchain?**  
   *Respuesta esperada:* El costo computacional (miner√≠a), la distribuci√≥n de copias en miles de nodos, y los mecanismos de consenso que requieren que la mayor√≠a de la red acuerde los cambios.

3. **¬øBlockchain es apropiada para todos los casos de uso?**  
   *Respuesta esperada:* No. Blockchain es √∫til cuando necesitas: descentralizaci√≥n, inmutabilidad, transparencia y desconfianza entre participantes. Para aplicaciones centralizadas o que requieren alta velocidad, bases de datos tradicionales son m√°s eficientes.

## üíª Actividad Pr√°ctica: Blockchain Completa

Vamos a crear una blockchain funcional que integre todo lo aprendido.

```python
import hashlib
import json
from datetime import datetime

class Blockchain:
    """
    Clase que representa una blockchain completa.
    """
    
    def __init__(self):
        """Inicializa la blockchain con el bloque g√©nesis."""
        self.chain = []
        self.crear_bloque_genesis()
    
    def crear_bloque_genesis(self):
        """Crea el primer bloque de la cadena."""
        bloque_genesis = self.crear_bloque(
            data='Bloque G√©nesis - Inicio de la Blockchain',
            previous_hash='0'
        )
        self.chain.append(bloque_genesis)
    
    def crear_bloque(self, data, previous_hash, nonce=0):
        """
        Crea un nuevo bloque con todos los campos necesarios.
        
        Par√°metros:
        - data: informaci√≥n a almacenar en el bloque
        - previous_hash: hash del bloque anterior
        - nonce: n√∫mero usado en miner√≠a (por defecto 0)
        
        Retorna: diccionario con el bloque completo
        """
        bloque = {
            'index': len(self.chain),  # √çndice autom√°tico basado en longitud de la cadena
            'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            'data': data,
            'previous_hash': previous_hash,
            'nonce': nonce,
            'hash': ''  # Se calcular√° a continuaci√≥n
        }
        bloque['hash'] = self.calcular_hash(bloque)
        return bloque
    
    def calcular_hash(self, bloque):
        """
        Calcula el hash SHA-256 de un bloque.
        
        Par√°metros:
        - bloque: diccionario con los campos del bloque
        
        Retorna: hash hexadecimal
        """
        # Excluir el campo 'hash' del c√°lculo
        bloque_sin_hash = {k: v for k, v in bloque.items() if k != 'hash'}
        # Convertir a JSON con claves ordenadas para consistencia
        bloque_string = json.dumps(bloque_sin_hash, sort_keys=True)
        # Calcular y retornar el hash
        return hashlib.sha256(bloque_string.encode('utf-8')).hexdigest()
    
    def agregar_bloque(self, data, nonce=0):
        """
        Agrega un nuevo bloque a la cadena.
        
        Par√°metros:
        - data: informaci√≥n a almacenar
        - nonce: n√∫mero usado en miner√≠a
        """
        # Obtener el hash del √∫ltimo bloque
        ultimo_bloque = self.chain[-1]
        previous_hash = ultimo_bloque['hash']
        
        # Crear y agregar el nuevo bloque
        nuevo_bloque = self.crear_bloque(data, previous_hash, nonce)
        self.chain.append(nuevo_bloque)
        
        print(f"‚úÖ Bloque #{nuevo_bloque['index']} agregado exitosamente")
    
    def mostrar_cadena(self):
        """Muestra todos los bloques de la cadena de forma legible."""
        print("\n" + "=" * 70)
        print(f"BLOCKCHAIN COMPLETA ({len(self.chain)} bloques)")
        print("=" * 70)
        
        for bloque in self.chain:
            print(f"\nüì¶ BLOQUE #{bloque['index']}")
            print("-" * 70)
            print(f"  Timestamp:     {bloque['timestamp']}")
            print(f"  Data:          {bloque['data']}")
            print(f"  Nonce:         {bloque['nonce']}")
            print(f"  Previous Hash: {bloque['previous_hash'][:16]}...")
            print(f"  Hash:          {bloque['hash'][:16]}...")
            print("-" * 70)
        
        print("=" * 70)

# Crear una blockchain
mi_blockchain = Blockchain()

# Agregar algunos bloques
mi_blockchain.agregar_bloque('Alice env√≠a 50 BTC a Bob')
mi_blockchain.agregar_bloque('Bob env√≠a 25 BTC a Charlie')
mi_blockchain.agregar_bloque('Charlie env√≠a 10 BTC a Diana')

# Mostrar la cadena completa
mi_blockchain.mostrar_cadena()
```

### üîç Explicaci√≥n del C√≥digo

1. **Clase `Blockchain`**  
   Encapsula toda la l√≥gica de la blockchain en una clase reutilizable.

2. **`__init__()`**  
   Constructor que inicializa la cadena vac√≠a y crea autom√°ticamente el bloque g√©nesis.

3. **`agregar_bloque()`**  
   M√©todo conveniente que obtiene autom√°ticamente el hash del √∫ltimo bloque y crea el enlace.

4. **`mostrar_cadena()`**  
   Visualiza toda la blockchain de forma estructurada.

## üîê Funci√≥n de Validaci√≥n Robusta

Ahora implementemos una funci√≥n que valide la integridad de toda la cadena.

```python
def is_chain_valid(blockchain):
    """
    Valida la integridad de una blockchain completa.
    
    Verifica:
    1. Que cada bloque tenga el hash correcto
    2. Que el previous_hash de cada bloque coincida con el hash del anterior
    3. Que el bloque g√©nesis tenga previous_hash = '0'
    
    Par√°metros:
    - blockchain: objeto Blockchain a validar
    
    Retorna: True si es v√°lida, False si est√° corrupta
    """
    chain = blockchain.chain
    
    print("\nüîç INICIANDO VALIDACI√ìN DE BLOCKCHAIN...")
    print("=" * 70)
    
    # Paso 1: Validar bloque g√©nesis
    if chain[0]['previous_hash'] != '0':
        print("‚ùå ERROR: El bloque g√©nesis no tiene previous_hash = '0'")
        return False
    print("‚úÖ Bloque g√©nesis v√°lido")
    
    # Paso 2: Validar cada bloque
    for i in range(len(chain)):
        bloque_actual = chain[i]
        
        # Verificar que el hash del bloque sea correcto
        hash_calculado = blockchain.calcular_hash(bloque_actual)
        if bloque_actual['hash'] != hash_calculado:
            print(f"‚ùå ERROR en Bloque {i}: Hash ha sido alterado")
            print(f"   Hash almacenado: {bloque_actual['hash'][:16]}...")
            print(f"   Hash calculado:  {hash_calculado[:16]}...")
            return False
        
        # Verificar encadenamiento (excepto para el bloque g√©nesis)
        if i > 0:
            bloque_anterior = chain[i - 1]
            if bloque_actual['previous_hash'] != bloque_anterior['hash']:
                print(f"‚ùå ERROR en Bloque {i}: Cadena rota")
                print(f"   Previous_hash del bloque: {bloque_actual['previous_hash'][:16]}...")
                print(f"   Hash del bloque anterior: {bloque_anterior['hash'][:16]}...")
                return False
        
        print(f"‚úÖ Bloque {i} v√°lido")
    
    print("=" * 70)
    print("‚úÖ BLOCKCHAIN COMPLETAMENTE V√ÅLIDA")
    print("=" * 70)
    return True

# Validar la blockchain
is_chain_valid(mi_blockchain)
```
